# Building a Mosaic

```python
import os
import cv2 

from PIL import Image
from skimage.io import imread
from skimage.io import imsave
from skimage import img_as_float
from tensorflow.keras.preprocessing.image import save_img
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
import photomosaic as pm
```

## Dog Mosaic

```python
def dog_mosaic(directory):
    #key image
    d_image = mpimg.imread(directory)
    d_image = img_as_float(d_image)
    
    #pool of images for mosaic tiles
    d_pool = pm.make_pool('../Data/folders/train/dog/*.jpg')
    
    #mosaic of 250x250 tiles
    d_mos = pm.basic_mosaic(d_image, d_pool, (250, 250), depth = 1)
    
    #save image to preferred location
    save_img('../images/retreiver.jpg', d_mos)
    
    #return mosaic
    return plt.imshow(d_mos)
```

```python
dog_mosaic('../images/key_image/unsplash_1.jpg');
```

## Cat Mosaic

```python
def cat_mosaic(directory):
    #key image
    c_image = mpimg.imread(directory)
    c_image = img_as_float(c_image)
    
    #pool of images for mosaic tiles
    c_pool = pm.make_pool('../Data/folders/train/cat/*.jpg')
    
    #mosaic of 250x250 tiles
    c_mos = pm.basic_mosaic(c_image, c_pool, (250, 250), depth = 1)
    
    #save image to preferred location
    save_img('../images/kitten.tif', c_mos)

    #return mosaic
    return plt.imshow(c_mos)
```

```python
cat_mosaic('../images/key_image/kitten.jpg');
```
